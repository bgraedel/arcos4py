## Imports

```python
from skimage import io
import numpy as np
from arcos4py.tools import track_events_image
import matplotlib.pyplot as plt
```

## Use track_events_image to track objects in a binary image

```python
img1 = io.imread("sample_data/pix/2_crossing.tif")
img2 = io.imread("sample_data/pix/4_colliding.tif")
```

```python
img_tracked = track_events_image(np.where(img1 > 0, 0, 1), 1.5)
```

## show the time series

```python
fig, axes = plt.subplots(4,4, figsize=(10, 10))
axes = axes.flatten()
for idx, i in enumerate(img_tracked):
    axes[idx].imshow(i)
    axes[idx].axis("off")
    axes[idx].set_title(f"t={idx}")

plt.show()
```

```python
img_tracked = track_events_image(np.where(img2 > 0, 0, 1), 1.5)
```
